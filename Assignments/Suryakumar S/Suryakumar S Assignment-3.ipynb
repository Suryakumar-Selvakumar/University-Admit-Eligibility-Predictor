### 4. Finding Outliers and removing

```python
sns.boxplot(x=data.Age)
```

```python
outlier = data.quantile((0.25,0.75))
outlier
```

```python
q1 = data.Age.quantile((0.25))
q1
```

```python
q3 = data.Age.quantile(0.75)
q3
```

```python
data.shape
```

```python
IQR = q3-q1
lower_limit = q1-1.5*IQR
upper_limit = q3+1.5*IQR
print(lower_limit, upper_limit)
```

#### so before removing outliers

```python
sns.boxplot(x=data.Age)
```

```python
# data1 = data[(data['Age'] > lower_limit) & (data['Age'] < upper_limit)]
data['Age'] = np.where(data['Age'] < lower_limit, 4, data['Age'])
data['Age'] = np.where(data['Age'] > upper_limit, 15, data['Age'])
data.shape
```

### 5. Checking Categorical data

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

data['Sex'] = le.fit_transform(data.iloc[:,0])
data['Sex'].unique()
```

```python
data.sample(5)
```

### 5. Identifying the dependent ans independent variable

```python
#dependent variable
y = data.Sex
y.head()
```

```python
X = data.drop(columns="Sex",axis=1)    #here if axis = 0 means row and axis = 1 means column....
X.head()
```

### 6. Scaling

```python
from sklearn.preprocessing import scale
sc_X = pd.DataFrame(scale(X), columns=X.columns)
sc_X.head()
```

### 7. Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sc_X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape,X_test.shape
```

### 8. Building the model

```python
from sklearn.ensemble import RandomForestClassifier as rfc
model = rfc(n_estimators=40,criterion='entropy')
```

```python
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
predict_train = model.predict(X_train)
```

### 9. Training and Testing accuracy

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print('Training accuracy: ',accuracy_score(y_train,predict_train))
print('Testing accuracy: ',accuracy_score(y_test,predict))
```

### 10. Measuring the performance

```python
pd.crosstab(y_test,predict)
```

```python
print(classification_report(y_test,predict))
```
